```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.signal as signal

from scipy.fft import fft, fftfreq
```

```python
accel_data = pd.read_csv("data/Biceps/accelerometer.csv", names=["timestamp", "x", "y", "z"])
gyro_data = pd.read_csv("data/Biceps/gyroscope.csv", names=["timestamp", "x", "y", "z"])
```

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(16, 10), constrained_layout=True)

for i, data in enumerate([accel_data, gyro_data]):
    for j, axis in enumerate(["x", "y", "z"]):
        axs[j, i].plot(data["timestamp"].values, data[axis].values)
        axs[j, i].set_title(f"{'Accelerometer' if i == 0 else 'Gyroscope'} - {axis.upper()} axis")
        axs[j, i].set_xlabel("Time (ms)")
        axs[j, i].set_ylabel("Value")
        axs[j, i].grid()
```

```python
signal_x = accel_data['x']  # example
time_x = accel_data['timestamp'] / 1000  # convert ms → s
```

```python
signal_z = gyro_data['z']  # example
time_z = gyro_data['timestamp'] / 1000  # convert ms → s
```

```python
# Plot selected axis
plt.figure(figsize=(10,4))
plt.plot(time_x, signal_x)
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (g)')
plt.title('Accelerometer X-axis Signal (Selected Axis)')
plt.show()
```

```python
# Plot selected axis
plt.figure(figsize=(10,4))
plt.plot(time_z, signal_z)
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (g)')
plt.title('Gyroscope Z-axis Signal (Selected Axis)')
plt.show()
```

```python
import numpy as np

sampling_intervals_x = np.diff(time_x)
sampling_intervals_z = np.diff(time_z)
sampling_rate_x = 1 / np.mean(sampling_intervals_x)
sampling_rate_z = 1 / np.mean(sampling_intervals_z)
print(f"Estimated Sampling Rate X: {sampling_rate_x:.2f} Hz")
print(f"Estimated Sampling Rate Z: {sampling_rate_z:.2f} Hz")
```

```python
from scipy.fft import fft, fftfreq

# Remove DC offset (mean)
signal_centered = signal_x - np.mean(signal_x)

N = len(signal_centered)
yf = fft(signal_centered)
xf = fftfreq(N, 1/sampling_rate)

# Plot frequency spectrum (positive frequencies only)
plt.figure(figsize=(10,4))
pos_mask = xf > 0
plt.plot(xf[pos_mask], np.abs(yf[pos_mask]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Frequency Spectrum - Accelerometer X-axis')
plt.show()
```
